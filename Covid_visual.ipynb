COVID DATA VISUALIZATION (PAKISTAN)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cd "D:\Data Science"
```

```python
covid_confirm = pd.read_csv('CONVENIENT_global_confirmed_cases.csv')
```

```python
for i in covid_confirm:
    if i != 'Country/Region' and i != 'Pakistan':
        covid_confirm.drop(i, axis = 1, inplace = True)
```

```python
covid_confirm.head()
```

```python
covid_confirm.drop(covid_confirm.index[0], inplace = True)
```

```python
covid_confirm.columns = ['Period', 'Cases']
```

```python
covid_confirm.head()
```

```python
covid_dead = pd.read_csv('CONVENIENT_global_deaths.csv')
```

```python
for i in covid_dead:
    if i != 'Country/Region' and i != 'Pakistan':
        covid_dead.drop(i, axis = 1, inplace = True)
```

```python
covid_dead.drop(covid_dead.index[0], inplace = True)
```

```python
covid_dead.columns = ['Period', 'Deaths']
```

```python
mega = pd.merge(covid_confirm, covid_dead)
```

```python
mega.head()
```

```python
mega.plot(x = 'Period', y = 'Cases', grid = True, color = 'g', alpha = 1, linewidth = 1)
plt.title('COVID Cases (Pakistan)')
plt.rcParams["figure.figsize"] = (50,3)

plt.show()
```

```python
deaths = mega.plot(x = 'Period', y = 'Deaths', grid = True, color = 'blue', alpha = 1, linewidth = 1)
deaths.legend(loc = 'upper right')
plt.title('COVID Deaths (Pakistan)')
plt.rcParams["figure.figsize"] = (20,4)

plt.show()
```
