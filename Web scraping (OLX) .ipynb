```python
from requests import get
url = "https://www.olx.com.pk/cars_c84"
response = get(url)

print(response.text[:500])
```

```python
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
```

```python
Vehicle_container = html_soup.find_all('div', class_ = "IKo3_")
print(type(Vehicle_container))
print(len(Vehicle_container))
```

```python
First_vehicle = Vehicle_container[0]
```

```python
price = First_vehicle.find('span', class_ = '_89yzn').text
```

```python
Name = First_vehicle.find('span', class_ = '_2tW1I').text
```

```python
Driven = First_vehicle.find('span', class_ = '_2TVI3').text
```

```python
Names = []
Prices = []
Driven = []

for container in Vehicle_container:
    name = container.find('span', class_ = '_2tW1I').text
    Names.append(name)
    
    price = container.find('span', class_ = '_89yzn').text
    Prices.append(price)
    
    driven = container.find('span', class_ = '_2TVI3').text
    Driven.append(driven)
```

```python
import pandas as pd

Vehicle_dataset = pd.DataFrame({'Name': Names, 
                               'Prices': Prices,
                               'Driven': Driven})

```

```python
Vehicle_dataset
```

```python
Vehicle_dataset.style.set_properties(**{'text-align': 'left'}, inplace = True)
```

```python
Vehicle_dataset.Prices[0].split()[1]
```

```python
Vehicle_dataset['Prices'] = Vehicle_dataset['Prices'].str.replace('Rs','')
```

```python
Vehicle_dataset['Prices'] = Vehicle_dataset['Prices'].str.replace(',','')
```

```python
Vehicle_dataset['Prices'][0]
```

```python
n = []
for i in Vehicle_dataset['Prices']:
    i = int(i)
    n.append(i)
```

```python
Vehicle_dataset['Prices'] = n
```

```python
Year = []
for i in Vehicle_dataset['Driven']:
    j = i.split('-')[0]
    Year.append(j)
    
    
    
```
